#### HANDLING THE OUTLIERS :-
* Outliers means "Extreme Values" on both the sides of the sorted data.
* Represented/detected/treated through BOX PLOT.
* Outliers can be handled through 2 ways:
        1. Outlier Detection:
                        i. %tile method
                        ii. IQR method
        2. Treatment:
                    .clip()

##### Q: Find the outliers in the TotalSales Column.

```python
# Handling Outliers using IQR Method :-
p0 = stores.TotalSales.min()
p0
```

```python
p25 = stores.TotalSales.quantile(0.25)
p25
```

```python
p50 = stores.TotalSales.quantile(0.50)
p50
```

```python
p75 = stores.TotalSales.quantile(0.75)
p75
```

```python
p100 = stores.TotalSales.max()
p100
```

```python
print(p0, p25, p50, p75, p100)
```

```python
# find iqr :-
iqr = p75 - p25
iqr
```

```python
# calculate the lower cut off :-
lc = p25 - 1.5 * iqr
lc
```

```python
# calculate the upper cut off :-
uc = p75 + 1.5 * iqr
uc
```

```python
print('p0 =', p0)
print('lc =', lc)
print('p100 =', p100)
print('uc =', uc)
```

**From the above cell, we can make the following conclusions :-** <br/>
1. lower cutoff lies way below minimum.
2. upper cutoff lies way above maximum.
3. If lower cutoff and upper cutoff doesn't lie in the range of MIN and MAX than Outliers can be ignored.
Or simply we can say that there are no Outliers in the **TotalSales** Column.

#### BOX PLOT :-

```python
stores.TotalSales.plot(kind = 'box')
```

It is clear from the BoxPlot that there are no outliers present in the TotalSales column.

##### Q: Find the outliers in the Total_Customers Column.

```python
p0 = stores.Total_Customers.min()
p25 = stores.Total_Customers.quantile(0.25)
p50 = stores.Total_Customers.quantile(0.50)
p75 = stores.Total_Customers.quantile(0.75)
p100 = stores.Total_Customers.max()
```

```python
print(p0, p25, p50, p75, p100)
```

```python
# calculating iqr :-
iqr = p75 - p25
iqr
```

```python
# calculating upper cut off :-
uc = p75 + 1.5 * iqr
uc
```

```python
# calculating lower cuttoff :-
lc = p25 - 1.5 * iqr
lc
```

```python
print('p0 =', p0)
print('lc =', lc)
print('p100 =', p100)
print('uc =', uc)
```

**From the above cell, we can make the following conclusions :-** <br/>
1. lower cutoff lies way below minimum. So, no outlier on lower side.
2. upper cutoff is less than maximum.So, outlier is present on upper side.
3. p100 = 335 <br/>
uc = 305.25 <br/>
What it basically means is area between 305.25 and 335  is Extreme Values or Outlier, and it should be treated accordingly.

#### BOX PLOT :-

```python
stores.Total_Customers.plot(kind = 'box')
```

We can see an outlier on the upper side of the BoxPlot.

##### Treatment of the outlier :- 
wherever the values are more than the uc, they will be replaced by the uc itself.

```python
stores.Total_Customers.clip(lower = lc, upper = uc)
```
